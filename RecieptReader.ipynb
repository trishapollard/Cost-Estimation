```python
import requests
import pandas as pd
import json

print("=== Python Receipt OCR Demo - Need help? Email support@asprise.com ===")

receiptOcrEndpoint = 'https://ocr.asprise.com/api/v1/receipt'  # Receipt OCR API endpoint
imageFile = "/content/receipt.jpg"  # Modify this to use your own file if necessary

# Sending the POST request to the OCR API
r = requests.post(receiptOcrEndpoint, data={
    'client_id': 'TEST',        # Use 'TEST' for testing purposes
    'recognizer': 'auto',       # can be 'US', 'CA', 'JP', 'SG' or 'auto'
    'ref_no': 'ocr_python_123',  # optional caller-provided ref code
}, files={"file": open(imageFile, "rb")})

# Parsing the result into JSON
result_json = r.json()

# Convert the OCR result into a DataFrame
df = pd.json_normalize(result_json)  # Normalize the nested JSON structure

# If the 'receipts' column has more nested info, unpack it
if 'receipts' in df.columns:
    receipts_df = pd.json_normalize(df['receipts'][0])  # Unpack the first receipt's info

    # Concatenate the receipts' detailed info with the original df
    df = pd.concat([df.drop(columns=['receipts']), receipts_df], axis=1)

# Display the DataFrame and its columns
print(df)
print(df.columns)
```
